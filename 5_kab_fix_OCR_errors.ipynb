```python
OCR_FIXES = {
    "фз": "фэ",
    "гц": "щ",
    "щг": "щт",
    "ьI": "ы",
    "ьш": "ым",
    "щI": "шI",
    "шц": "ищ",
    "Ю": "Iо",
    "IД": "Щ",
    "кГ": "кI",
    "цГ": "цI",
    "пгь": "шъ",
    "гь": "гъ",
    "ээ": "зэ",
    "щъ": "шъ",
    "ГI": "П",
    "гI": "тI",
    "IЦ": "Щ",
    "еы": "сы",
    "кь": "къ",

    "ьпп": "ыш",
    "ьпц": "ыщ",
    "пгг": "шт",
    "пг": "ш",
    "нн": "ин",
    "рн": "ри",
    "хн": "хи",
    "ьнц": "ыщ",
    "ьн": "ьи",
    "III": "Ш",
    "мн": "ми",
    "нп": "ип",
    "сн": "си",
    "лн": "ли",
    "ън": "ъи",

    "Лр": "Ар",
    "Лщ": "Ащ",
    "шь": "шъ",
    "ге": "те",

    "зз": "зэ",
    "дI": "цI",
    "кIп": "кIи",
    "рп": "ри",
    "гх": "тх",
    "зьг": "зы",
    "еэ": "сэ",
    "ль": "лъ",

    "ъI": "ы",
    "нх": "их",
    "чь": "чъ",
    "гк": "тк",
    "еи": "си",
    "ии": "ин",
    "нм": "им",
    "фн": "фи",
    "дн": "ди",
    "Iн": "Iи",

    "тъ": "гъ",
    "пь": "гъ",
    "иахь": "нахь",

    "шГ": "шI",
    "шты": "щты"
}
```

```python
input_file = "/content/drive/MyDrive/4_Kabard_NoOCR_balanced.txt"
output_file = "/content/drive/MyDrive/4_Kabard_NoOCR_fixed_balanced.txt"
```

```python
with open(input_file, "r", encoding="utf-8") as f:
    text = f.read()

# нормализация палочки
text = text.replace("Ӏ", "I")

for wrong in sorted(OCR_FIXES, key=len, reverse=True):
    text = text.replace(wrong, OCR_FIXES[wrong])

with open(output_file, "w", encoding="utf-8") as f:
    f.write(text)

print("Готово:", output_file)
```
