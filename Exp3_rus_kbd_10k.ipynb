```python
import os
import shutil
import re

img_dir = "/content/kab_img"
gt_dir = "/content/kab_gt"

pattern = re.compile(r"_(\d{6})_")

for img in os.listdir(img_dir):

    if not img.endswith(".jpg"):
        continue

    num = int(pattern.search(img).group(1))

    if num <= 90000:
        split = "train"
    elif num <= 95000:
        split = "valid"
    else:
        split = "test"

    img_src = os.path.join(img_dir, img)
    gt_src = os.path.join(gt_dir, img.replace(".jpg",".gt.txt"))

    img_dst = f"{base}/{split}/images/{img}"
    gt_dst = f"{base}/{split}/gt/{img.replace('.jpg','.gt.txt')}"

    shutil.move(img_src, img_dst)
    shutil.move(gt_src, gt_dst)
```

```python
import os
import re
import random

def reduce_renders_per_text(img_dir, gt_dir, keep=4):

    random.seed(42)

    pattern = re.compile(r"_(\d{6})_")
    groups = {}

    for img in sorted(os.listdir(img_dir)):

        if not img.endswith(".jpg"):
            continue

        m = pattern.search(img)
        if not m:
            continue

        text_id = m.group(1)
        groups.setdefault(text_id, []).append(img)

    removed = 0

    for text_id in sorted(groups.keys()):

        imgs = groups[text_id]

        if len(imgs) <= keep:
            continue

        keep_imgs = set(random.sample(imgs, keep))

        for img in imgs:

            if img not in keep_imgs:

                os.remove(os.path.join(img_dir, img))
                os.remove(os.path.join(gt_dir, img.replace(".jpg",".gt.txt")))

                removed += 1

    print("Удалено:", removed)


reduce_renders_per_text(
"/content/kab_dataset/train/images",
"/content/kab_dataset/train/gt",
1
)

reduce_renders_per_text(
"/content/kab_dataset/valid/images",
"/content/kab_dataset/valid/gt",
1
)

reduce_renders_per_text(
"/content/kab_dataset/test/images",
"/content/kab_dataset/test/gt",
1
)
```

```python
import subprocess
import pandas as pd
from jiwer import cer, wer
from tqdm import tqdm

def evaluate_ocr(img_dir, gt_dir, model, output_csv):

    results = []
    files = os.listdir(img_dir)

    for img in tqdm(files):

        img_path = os.path.join(img_dir, img)
        out_path = "/content/tmp_out"

        subprocess.run([
            "tesseract",
            img_path,
            out_path,
            "-l",
            model,
            "--psm",
            "7"
        ], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

        with open(out_path + ".txt") as f:
            pred = f.read().strip()

        with open(os.path.join(gt_dir, img.replace(".jpg",".gt.txt"))) as f:
            gt = f.read().strip()

        gt = " ".join(gt.split())
        pred = " ".join(pred.split())

        results.append({
            "file": img,
            "gt": gt,
            "prediction": pred,
            "cer": cer(gt, pred),
            "wer": wer(gt, pred)
        })

    df = pd.DataFrame(results)

    df.to_csv(output_csv, index=False)

    print("Saved:", output_csv)
    print("CER:", df["cer"].mean())
    print("WER:", df["wer"].mean())

    return df
```

```python
import os
import shutil
from tqdm import tqdm
import random

train_src_img = "/content/kab_dataset/train/images"
train_src_gt  = "/content/kab_dataset/train/gt"

valid_src_img = "/content/kab_dataset/valid/images"
valid_src_gt  = "/content/kab_dataset/valid/gt"

train_dst = "/content/tesstrain/data/rus_kbd-ground-truth"
valid_dst = "/content/tesstrain/data/rus_kbd-eval"

os.makedirs(train_dst, exist_ok=True)
os.makedirs(valid_dst, exist_ok=True)

random.seed(42)

files = [f for f in os.listdir(train_src_img) if f.endswith(".jpg")]
files = random.sample(files, 10000)

for f in tqdm(files):

    shutil.copy(
        os.path.join(train_src_img, f),
        os.path.join(train_dst, f)
    )

    shutil.copy(
        os.path.join(train_src_gt, f.replace(".jpg",".gt.txt")),
        os.path.join(train_dst, f.replace(".jpg",".gt.txt"))
    )

print("Copy validation data")

for f in tqdm(os.listdir(valid_src_img)):

    shutil.copy(
        os.path.join(valid_src_img, f),
        os.path.join(valid_dst, f)
    )

    shutil.copy(
        os.path.join(valid_src_gt, f.replace(".jpg",".gt.txt")),
        os.path.join(valid_dst, f.replace(".jpg",".gt.txt"))
    )
```

```python
from PIL import Image
import os
from tqdm import tqdm

def convert_and_fix(folder):

    converted = 0

    for f in tqdm(os.listdir(folder)):

        path = os.path.join(folder, f)

        if f.endswith(".jpg"):

            img = Image.open(path)
            png_path = path.replace(".jpg", ".png")

            img.save(png_path, dpi=(300,300))
            os.remove(path)

            converted += 1

        elif f.endswith(".png"):

            img = Image.open(path)
            img.save(path, dpi=(300,300))

    print("converted jpg → png:", converted)


convert_and_fix("/content/tesstrain/data/rus_kbd-ground-truth")
convert_and_fix("/content/tesstrain/data/rus_kbd-eval")
```

```python
import shutil
import os
import time
import threading

src_dir = "/content/tesstrain/data"
dst_dir = "/content/drive/MyDrive/OCR_models/checkpoints"

os.makedirs(dst_dir, exist_ok=True)

def autosave():
    while True:
        for f in os.listdir(src_dir):
            if "rus_kbd_checkpoint" in f:
                shutil.copy(
                    os.path.join(src_dir, f),
                    os.path.join(dst_dir, f)
                )
                print("Saved:", f)

        time.sleep(300)  # каждые 5 минут

thread = threading.Thread(target=autosave)
thread.daemon = True
thread.start()
```
