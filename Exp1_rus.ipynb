Оставляем только 10000 примеров в трейн и 5000 валид и 5000 тест

```python
import os
import re
import random

def reduce_renders_per_text(img_dir, gt_dir, keep=4):

    random.seed(42)

    pattern = re.compile(r"_(\d{6})_")
    groups = {}

    for img in sorted(os.listdir(img_dir)):

        if not img.endswith(".jpg"):
            continue

        m = pattern.search(img)
        if not m:
            continue

        text_id = m.group(1)
        groups.setdefault(text_id, []).append(img)

    removed = 0

    for text_id in sorted(groups.keys()):

        imgs = groups[text_id]

        if len(imgs) <= keep:
            continue

        keep_imgs = set(random.sample(imgs, keep))

        for img in imgs:

            if img not in keep_imgs:

                os.remove(os.path.join(img_dir, img))
                os.remove(os.path.join(gt_dir, img.replace(".jpg",".gt.txt")))

                removed += 1

    print("Удалено:", removed)


reduce_renders_per_text(
"/content/kab_dataset/train/images",
"/content/kab_dataset/train/gt",
1
)

reduce_renders_per_text(
"/content/kab_dataset/valid/images",
"/content/kab_dataset/valid/gt",
1
)

reduce_renders_per_text(
"/content/kab_dataset/test/images",
"/content/kab_dataset/test/gt",
1
)
```

Комбинируем датасет из двух языков

```python
import os
import shutil
import random
from tqdm import tqdm

random.seed(42)  # фиксируем выборку для всех участников

base = "/content/combined_dataset"

for split in ["train","valid","test"]:
    os.makedirs(f"{base}/{split}/images", exist_ok=True)
    os.makedirs(f"{base}/{split}/gt", exist_ok=True)

# сколько брать из каждого языка
limits = {
    "train": 5000,
    "valid": 2500,
    "test": 2500
}

def sample_and_copy(src_base, dst_base, split, limit, prefix):

    imgs = sorted(os.listdir(f"{src_base}/{split}/images"))

    # фиксированная выборка
    selected = random.sample(imgs, limit)

    for img in tqdm(selected):

        # добавляем префикс, чтобы не было перезаписи
        new_img_name = f"{prefix}_{img}"
        new_gt_name = f"{prefix}_{img.replace('.jpg','.gt.txt')}"

        shutil.copy(
            f"{src_base}/{split}/images/{img}",
            f"{dst_base}/{split}/images/{new_img_name}"
        )

        shutil.copy(
            f"{src_base}/{split}/gt/{img.replace('.jpg','.gt.txt')}",
            f"{dst_base}/{split}/gt/{new_gt_name}"
        )

# собираем датасет
for split in ["train","valid","test"]:

    print(f"\nProcessing {split}")

    # адыгейский
    sample_and_copy("/content/ady_dataset", base, split, limits[split], "ady")

    # кабардинский
    sample_and_copy("/content/kab_dataset", base, split, limits[split], "kab")
```

```python
import zipfile
import os
from tqdm import tqdm

def zip_folder(src, dst):

    with zipfile.ZipFile(dst, 'w', zipfile.ZIP_DEFLATED) as z:

        for root, _, files in os.walk(src):
            for file in tqdm(files):
                full = os.path.join(root, file)
                rel = os.path.relpath(full, src)
                z.write(full, rel)
```

```python
from jiwer import cer, wer
import pandas as pd
import subprocess
from tqdm import tqdm
import os

def evaluate(img_dir, gt_dir, model, output_csv):

    results = []

    for img in tqdm(os.listdir(img_dir)):

        if not img.endswith(".jpg"):
            continue

        img_path = os.path.join(img_dir, img)
        out_path = "/content/tmp_out"

        subprocess.run([
            "tesseract",
            img_path,
            out_path,
            "-l", model,
            "--psm", "7"
        ], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

        with open(out_path + ".txt") as f:
            pred = f.read().strip()

        with open(os.path.join(gt_dir, img.replace(".jpg",".gt.txt"))) as f:
            gt = f.read().strip()

        results.append({
            "file": img,
            "gt": gt,
            "prediction": pred,
            "cer": cer(gt, pred),
            "wer": wer(gt, pred)
        })

    df = pd.DataFrame(results)
    df.to_csv(output_csv, index=False)

    print("Saved:", output_csv)
    print("CER:", df["cer"].mean())
    print("WER:", df["wer"].mean())

    return df
```
